```python
df = pd.read_csv("./survey_results_responses.csv")

df.columns
```

```python
responses = df['ResponseId']
languages = df['LanguageHaveWorkedWith']
```

```python
responses.head(5)
```

```python
from collections import Counter
```

```python
languages_counter = Counter()

for lang in languages:
    if isinstance(lang, str):
        languages_counter.update(lang.split(';'))    
```

```python
languages = []
lang_popularity = []

for item in languages_counter.most_common(15):
    languages.append(item[0])
    lang_popularity.append(item[1])

print(languages)    
print(lang_popularity)
```

```python
languages.reverse()
lang_popularity.reverse()

plt.barh(languages, lang_popularity)
```
