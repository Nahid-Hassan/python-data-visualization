```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ages = np.arange(25,36)
ages
```

```python
dev_avg = [38233,40002,45244,48242,52325,53325, 57515,62422,61433, 67002,72342]
```

```python
plt.plot(ages, dev_avg)
```

```python
plt.bar(ages, dev_avg)
```

```python
dev_avg = [38233,40002,45244,48242,52325,53325, 57515,62422,61433, 67002,72342]
py_dev_avg = [45423,47345,52423,56342,55423,57234,64242,67243,72243,76234,78234]

plt.bar(ages, py_dev_avg, label="Python Dev. Avg Salary")
plt.bar(ages, dev_avg, label="All Dev. Avg Salary")

plt.title('Developer Salary Survey')
plt.xlabel('Ages')
plt.ylabel('Salary')

plt.legend()
plt.show()
```
