```python
import pandas as pd
```

```python
people = {
    "first": ["Nahid", "Mehedi", "Mony"],
    "last" : ["Hassan", "Mahin", "Mia"],
    "email": ["nahid@gmail.com", "m12@gmail.com", "mia@gmail.com"]
}
```

```python
df = pd.DataFrame(people)

df.head()
```

### Sort Columns

```python
df.sort_values(by="first")
```

```python
df.sort_values(by="first", ascending=False)
```

```python
df.sort_values(by=['last', 'first'], ascending=False)
```

### Sort one column in ascending order and another is desending order

```python
people = {
    "first": ["Nahid", "Nahid", "Mony", "Adam", "Xen"],
    "last" : ["Hassan", "Mahin", "Mia", "Wad", "Yia"],
    "email": ["nahid@gmail.com", "m12@gmail.com", "mia@gmail.com", "ada@gmail.com", "xen@gmail.com"]
}

df = pd.DataFrame(people)
df
```

```python
df.sort_values(by=['last', 'first'], ascending=False)
```

```python
df.sort_values(by=['last', 'first'], ascending=[False, True])
```

```python
df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
```

```python
df
```

```python
df.sort_index()
```

```python
df['last'].sort_values()
```
